## Data Flattened

### Test mechanical properties

```python
def flatten_json_by_series(y):
    out = {}

    def flatten(x, name='', counter_dict={}):
        if isinstance(x, dict):
            for a in x:
                # Construimos nuevo nombre de la clave
                new_name = name + a
                # Comprobamos si el nombre ya existe en el diccionario de salida
                if new_name in counter_dict:
                    # Si ya existe, incrementamos el contador y actualizamos el nombre
                    counter_dict[new_name] += 1
                    current_name = new_name + '_' + str(counter_dict[new_name]) + '.'
                else:
                    # Si no existe, lo agregamos al diccionario con un contador inicial
                    counter_dict[new_name] = 0
                    current_name = new_name + '.'
                # Llamamos recursivamente a la funciÃ³n flatten
                flatten(x[a], current_name, counter_dict)
        elif isinstance(x, list):
            i = 0
            for a in x:
                flatten(a, name + str(i) + '.', counter_dict)
                i += 1
        else:
            # Ajustamos el nombre final quitando el punto o guiÃ³n bajo extra al final
            final_name = name.rstrip('.').rstrip('_0')
            out[final_name] = x

    flatten(y)
    return out

def apply_flatten(serie):
    # Aplica la funciÃ³n flatten_json a cada elemento de la Serie
    return serie.apply(flatten_json_by_series).apply(pd.Series)
```

```python
df_test_flattened = apply_flatten(df_all['data.ingredients'])
df_test_flattened
```

# Output pre-processing

```python
def flatten_json(y, root_name=''):
    out = {}

    def flatten(x, name=''):
        # Se usa el nombre de la raÃ­z proporcionado si estamos en el nivel superior
        full_name = f"{root_name}{name}" if name else root_name

        if isinstance(x, dict):
            for a in x:
                flatten(x[a], f"{name}{a}_")
        elif isinstance(x, list):
            for i, a in enumerate(x):
                flatten(a, f"{name}{i}.")
        else:
            # Ajuste para remover el separador final en el nombre de la clave
            final_name = full_name.rstrip('_').rstrip('.')
            out[final_name] = x

    flatten(y)
    return out

def apply_flatten_by_series(df, column_name):
    flattened_series = df[column_name].apply(lambda x: flatten_json(x, root_name=column_name+'_'))
    return pd.DataFrame(flattened_series.tolist())
```

```python
df_full_data = pd.DataFrame()
for key in df_all.columns:
    try:
        df_aux = df_all
    except:
        break
    df_aux1 = apply_flatten_by_series(df_aux, key)
    df_full_data = pd.concat([df_full_data, df_aux1], axis=1)
df_full_data
```
